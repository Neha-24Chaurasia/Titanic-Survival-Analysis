```python
# See from above table we see that ticket CA 2144 is assigned to all 8 members n all have same SibSp n Parch value n Fare value. So we will find out individual fare of all the passengers.

df['Individual_Fare']=df['Fare']/ (df['SibSp']+df['Parch']+1)
df['Individual_Fare']
```

```python
# Now doing Feature Engineering on SibSp and Parch
df['Family_size']=df['SibSp']+df['Parch']+1
df
```

```python
# Family_type
# 1 -> alone
# 2-4 -> small
# 5 0r more -> large
def transform_family_size(num):
 if num == 1:
    return 'alone'
 elif num>1 and num<5:
    return 'small'
 else:
    return 'large'
    
```

```python
df['Family_type']= df['Family_size'].apply( transform_family_size)
df
```

```python
# Performing bivariate Analysis on Family_type col
pd.crosstab(df['Survived'],df['Family_type'],normalize='columns')*100
```

```python
# Conclusion- If ur travelling alone than high chances i.e 69% chances that u wi  die if large family than there is 83% chances that u die n if small family than there is high chances u survive
```

```python
# Applying feature engineering on Name column
df['Title']=df['Name'].apply(lambda x:x.split(',')[1].split('.')[0].strip())
df['Title']
```

```python
# rare title ka ek group bna do bcoz model me bhot jyada categories nai rakhte n model gets confused seeing so many categories.
df['Title']=df['Title'].replace(['Dr', 'Rev', 'Col','Major','Lady', 'Sir','Don','Countess','Capt','Johnkheer','Dona'],'Rare')

# Do Mapping of Title bcoz model dont understand text
df['Title']=df['Title'].map({'Mr':1,'Miss':2, 'Mrs':3, 'Master': 4, 'Rare':5})

# Name ki length find out karo bcoz big name almost rich families me hote h n pclass1 k log jyada survive kiye bcoz they paid more so its possible that big name length are rich people
df['Name_length']= df['Name'].apply(len)

# Now as we have applied feature engineering on name col we will drop Nane col as its of no use now
df.drop('Name', axis=1,inplace=True)

```

```python
# Cabin columns has many missing values so first fill that and than extract 1st letter of cabin n it is deck eg. C85 so Cis deck n show which class passengers are on it so it helps us to find if first, 2nd or 3rd class passengers are there on it n who survived
df['Cabin'].isnull().sum()
```

```python
# Missing values fill
df['Cabin']=df['Cabin'].fillna('M')

# Ab 1st value of deck i.e C in C85 chahiye 
df['Deck']=df['Cabin'].apply(lambda x: x[0])

# Now transform deck to numeric value bcoz model understand numeric data
df['Deck'].fillna('M',inplace=True)
df['Deck']=df['Deck'].map({'A':1, 'B': 2, 'C': 3, 'D': 4, 'E':5, 'F':6, 'G':7, 'T': 8, 'M': 9})

# As we have extracted info from cabin col so drop that col now.
df.drop('Cabin', axis=1,inplace=True)
```

#### Applying logistic Regression model to see their accuracy 





```python
train_df=df[df['Survived'].notnull()]
test_df=df[df['Survived'].isnull()]
```

```python
X=train_df.drop('Survived',axis=1)
y=train_df['Survived']
```

```python
# Scaling 
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
# Train and Test Split
from sklearn.model_selection import train_test_split

X_train,X_val,y_train,y_val=train_test_split( X_scaled,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model= LogisticRegression(max_iter=1000)
```

```python
# Model Training
model.fit(X_train,y_train)
```

```python
# Prediction on Validation data ...X_train ka 20% part X_val me jayega randomly everytime
y_pred=model.predict(X_val)
```

```python
# Accuracy 
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_val,y_pred)
print("Accuracy:",accuracy*100)
```

```python
# Confusion Matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_val,y_pred)
print(cm)

```

```python
# Final Test Prediction
final_test=test_df.drop('Survived', axis=1)
final_test_scaled=scaler.transform(final_test)
predictions=model.predict(final_test_scaled)
print(predictions[:20])
```
